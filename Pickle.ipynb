```python
#import library
import pandas as pd

data_german = pd.read_csv('D:\german_credit.csv',delimiter=',') 

#menampilkan beberapa baris pertama datapilot
data_german.head()
```

```python
#memisahkan data variabel X dan variabel Y
y=data_german['Creditability']

x_categorical =data_german[['Account Balance', 'Value Savings/Stocks']]
x_numerical = data_german[['Credit Amount']]

x_ubah = pd.get_dummies(x_categorical[['Account Balance', 'Value Savings/Stocks']],drop_first=True)
x_all = pd.concat([x_ubah, x_numerical], axis = 1)
```

```python
#pembuatan model berdasarkan data training
from sklearn.tree import DecisionTreeClassifier
pohon = DecisionTreeClassifier()

pohon.fit(x_all, y)

prediksi = pohon.predict(x_all)
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y, prediksi))
print("Precision:",metrics.precision_score(y, prediksi))
print("Recall:",metrics.recall_score(y, prediksi))
print("Specificity:",metrics.recall_score(1-y, 1-prediksi))
```

```python
import pickle

# save the model to disk
filename = 'finalized_model.sav'
pickle.dump(pohon, open(filename, 'wb'))
 
# some time later...
```

```python
# load the model from disk
loaded_model = pickle.load(open('finalized_model.sav', 'rb'))
```

```python
result = loaded_model.score(x_all, y)
print(result)
```
